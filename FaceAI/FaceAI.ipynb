```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/Users/Starkjing/Desktop/ComputerVision/FaceAI/1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faceCascade = cv2.CascadeClassifier("/Users/Starkjing/Desktop/ComputerVision/FaceAI/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml")

faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.2,
    minNeighbors=5,
    minSize=(30, 30),
)
color = (0, 255, 0)  # :
if len(faces):  # 0
    for faceRect in faces:  # 
        x, y, w, h = faceRect
        # 
        cv2.rectangle(img, (x, y), (x + h, y + w), color, 2)
        # 
        cv2.circle(img, (x + w // 4, y + h // 4 + 30), min(w // 8, h // 8),
                   color)
        #
        cv2.circle(img, (x + 3 * w // 4, y + h // 4 + 30), min(w // 8, h // 8),
                   color)
        #
        cv2.rectangle(img, (x + 3 * w // 8, y + 3 * h // 4),
                      (x + 5 * w // 8, y + 7 * h // 8), color)
    
plt.figure('Girl')
plt.imshow(img)
#plt.axis('off') # if you want to remove coordinates
plt.show()
```
